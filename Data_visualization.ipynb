# Bike thefts are most common in Downtown Seattle in August, 2014

Figure A shows the regional distribution of bike thefts in three months overlaid with the matched latitude and longitude in the map of Seattle. From figure A, it is obvious that the downtown Seattle is the region with the highest density of data points. Figure B&C shows the number of bike theft in three months at Seattle and Seattle downtown (latitude within 47.58-47.65, longitude within -122.36 - -122.30). These two figures show that bike theft occurred more in July and August both at Seattle and in August at Seattle downtown. To conclude, Bike thefts are most common in Downtown Seattle in August, 2014.

```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd

with open('seattle_incidents_summer_2014.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    bike_thefts=[]
    for row in csv_reader:
        if line_count == 0:
            print(f'Column names are {", ".join(row)}')
            line_count += 1
        else:
            if row[4]=='THEFT-BICYCLE':
                  bike_thefts.append(row)
                  line_count += 1
    print(f'Processed {line_count} lines for bike_theft.')
```

```python
good_bike_thefts_record=[]
count=0
for bike_theft in bike_thefts:
    if (float(bike_theft[14])!= 0) and (float(bike_theft[15])!=0):
        good_bike_thefts_record.append(bike_theft)
        count =count+1
print(f'There are {count} records with valid altitude and longitude.')
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
img=mpimg.imread('Picture1.png')
imgplot = plt.imshow(img)
#img.show()
```

```python
count_6=0
count_7=0
count_8=0
for record in good_bike_thefts_record:
    if record[17] == '6':
        count_6=count_6+1
    elif record[17] == '7':
        count_7=count_7+1
    elif record[17] == '8':
        count_8=count_8+1
print('Numbers of bike thefts in Jun, Jul and Aug are {},{},{}'.format(count_6,count_7,count_8))
```
