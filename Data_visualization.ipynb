```python
import csv

with open('seattle_incidents_summer_2014.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    bike_thefts=[]
    for row in csv_reader:
        if line_count == 0:
            print(f'Column names are {", ".join(row)}')
            line_count += 1
        else:
            if row[4]=='THEFT-BICYCLE':
                  bike_thefts.append(row)
                  line_count += 1
    print(f'Processed {line_count} lines for bike_theft.')
```

```python
good_bike_thefts_record=[]
count=0
for bike_theft in bike_thefts:
    if (float(bike_theft[14])!= 0) and (float(bike_theft[15])!=0):
        good_bike_thefts_record.append(bike_theft)
        count =count+1
print(f'There are {count} records with valid altitude and longitude.')
```

```python
count_6=0
count_7=0
count_8=0
for record in good_bike_thefts_record:
    if record[17] == '6':
        count_6=count_6+1
    elif record[17] == '7':
        count_7=count_7+1
    elif record[17] == '8':
        count_8=count_8+1
print('Numbers of bike thefts in Jun, Jul and Aug are {},{},{}'.format(count_6,count_7,count_8))
```

```python
bike_theft_downtown=[]
count_dt_6=0
count_dt_7=0
count_dt_8=0
for bike_theft in good_bike_thefts_record:
    if (-122.36<float(bike_theft[14])<-122.3) and (47.58<float(bike_theft[15])<47.65):
        if bike_theft[17] == '6':
            count_dt_6=count_dt_6+1
        elif bike_theft[17] == '7':
            count_dt_7=count_dt_7+1
        elif bike_theft[17] == '8':
            count_dt_8=count_dt_8+1
print('Numbers of bike thefts at Downtown Seattle in Jun, Jul and Aug are {},{},{}'.format(count_dt_6,count_dt_7,count_dt_8))
```
